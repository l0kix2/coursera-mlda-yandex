```python
#импортируем все необходимое

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
import math
%matplotlib inline
```

```python
#задаем распределение

size = 1000
a=4
powerlaw_rv = sts.powerlaw(a)
```

```python
#создаем выборку

sample = powerlaw_rv.rvs(size)

#строим гистограмму

plt.hist(sample, normed=True, label = 'hist')

#поверх гисторграммы строим график теоретической плотности распределения
x = np.linspace(0,1,1000)
pdf = powerlaw_rv.pdf(x)
plt.plot(x, pdf, label='theoretical pdf', alpha=1)
plt.legend()
plt.ylabel('$f(x)$, fraction of samples')
plt.xlabel('$x$')
```

```python
#создаем матрицу, в которой содержатся наши выборки

n=5
sample_matrix = np.array([powerlaw_rv.rvs(n) for i in range(size)])

#считаем среднее для каждой выборки
means = np.array([sample_matrix[i].mean() for i in range (size)])

#строим гистограмму
plt.hist(means, normed=True, label = 'hist')

#поверх гисторграммы строим график теоретической плотности распределения, вычысляя теоретические значения
x = np.linspace(0.4 ,1.2,1000)
a = float(a)
m = a/(a+1)
d = math.sqrt((a/(a+2) - 2*a*a/((a+1) ** 2) + (a ** 2 / ((a+1) ** 2)))/n)
norm_rv = sts.norm(m, d)
pdf = norm_rv.pdf(x)
plt.plot(x, pdf, label='theoretical pdf', alpha=1)
plt.legend()
plt.ylabel('$f(x)$, fraction of samples')
plt.xlabel('$x$')
```
