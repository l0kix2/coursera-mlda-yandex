```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline

# Распределение Коши https://ru.wikipedia.org/wiki/%D0%A0%D0%B0%D1%81%D0%BF%D1%80%D0%B5%D0%B4%D0%B5%D0%BB%D0%B5%D0%BD%D0%B8%D0%B5_%D0%9A%D0%BE%D1%88%D0%B8

# Выборка объёма 1000 
cauchy_rv = sts.cauchy(0, 1)
sample = cauchy_rv.rvs(1000)

# Построение гистограммы выборки 
plt.hist(sample, bins=100, normed=True)
plt.ylabel('fraction of samples')
plt.xlabel('$x$')

# Построение теоретической плотности распределения поверх гистограммы выборки
x = np.linspace(-5,5,100)
pdf = cauchy_rv.pdf(x)
plt.plot(x, pdf)

plt.ylabel('$f(x)$')
plt.xlabel('$x$')

```

```python
# Распределение выборочного среднего при объёме выборки n равной 5 
# Расчёт теоретического мат ожидания
# E[X]=\int \limits _{{-\infty }}^{{\infty }}\!xf_{X}(x)\,dx
# Расчёт теоретической дисперсии исходного распределения
# {\displaystyle D[X]=\int _{-\infty }^{\infty }{f(x)(x-M[X])^{2}dx}}

from numpy import random, mean, var

xm = mean(sample)  # среднее значение
xv = var(sample)   # дисперсия

from numpy import random
r = random.normal(xm, xv, size=5)

# Построение гистограммы распределений их выборочных средних

plt.hist(r, bins=30, normed=True)
plt.ylabel('fraction of samples')
plt.xlabel('$x$')

# Построение графика плотности соответствующего нормального распределения
```
